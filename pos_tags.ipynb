```python
import pandas as pd
import spacy
```

```python
# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Load the corpus
from google.colab import files
import io
uploded = files.upload()
corpus = pd.read_csv(io.BytesIO(uploded['bbc-text.csv']))
corpus = corpus[:100000]
corpus = corpus[:100]  # For the sake of time, we'll only use the first 100 rows

def extract_pos_tags(text):
    doc = nlp(text)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags

# Apply the function to each row in the DataFrame
corpus['pos_tags'] = corpus['text'].apply(extract_pos_tags)

# Display the DataFrame with the new 'pos_tags' column
print(corpus.head())
```
